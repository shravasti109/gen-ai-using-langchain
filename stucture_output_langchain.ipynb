```
Behind the scenes, LangChain automatically generates a system prompt. This prompt tells the LLM that it should return the output in a specific format containing a summary and sentiment.

Then, the user's review text is sent to the LLM along with this generated system prompt. The model reads both and tries to produce the required structured output.

However, sometimes it can still be difficult for the LLM to understand exactly what we want. The model may misunderstand a field name or return incorrect data.

To help the LLM understand better, we can provide additional descriptions for each field. Giving extra information about a field is called annotation.

We use Annotated from the typing module:

from typing import TypedDict, Annotated

Example:

class Review(TypedDict):
    summary: Annotated[str, "A short summary of the review"]
    sentiment: Annotated[str, "The sentiment of the review: positive, negative, or neutral"]

Now the LLM receives more information about what each field means, making it more likely to return the correct structured output.
```

````python
from typing import TypedDict, Annotated
from langchain_google_genai import ChatGoogleGenerativeAI


class Review(TypedDict):
    summary: Annotated[
        str,
        "A short summary of the review"
    ]

    sentiment: Annotated[
        str,
        "Overall sentiment: positive, negative, or neutral"
    ]

    pros: Annotated[
        list[str],
        "List of positive points mentioned in the review"
    ]

    cons: Annotated[
        list[str],
        "List of negative points mentioned in the review"
    ]

    good_things: Annotated[
        list[str],
        "List of features or aspects that the customer liked"
    ]

    rating: Annotated[
        int,
        "Estimated rating from 1 to 5 based on the review"
    ]


model = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash"
)

structured_model = model.with_structured_output(Review)

review = """
I have been using this smartphone for about three weeks, and my overall experience has been quite positive. The camera quality is excellent, especially in daylight, and the pictures come out sharp with accurate colors. The battery life is one of the best features of this phone, as it easily lasts an entire day even with heavy usage. The display is bright, vibrant, and enjoyable for watching videos and playing games.

The performance is smooth for everyday tasks such as browsing, social media, video streaming, and multitasking. Applications open quickly, and I have not experienced any major lag. The design of the phone also feels premium, and it is comfortable to hold.

However, there are a few drawbacks. The device tends to become warm during long gaming sessions and while charging. The charging speed is acceptable but slower than some competing phones in the same price range. I also noticed that the camera performance in low-light conditions could be improved, as images sometimes appear noisy.

Despite these issues, I believe the phone offers great value for money. It provides a good balance of performance, battery life, display quality, and camera features. I would recommend it to students, professionals, and anyone looking for a reliable smartphone without spending too much money.
"""

result = structured_model.invoke(review)

print(result)
````

````python
from pydantic import BaseModel, Field
from langchain_google_genai import ChatGoogleGenerativeAI


class Review(BaseModel):
    summary: str = Field(
        description="A short summary of the review"
    )

    sentiment: str = Field(
        description="Overall sentiment: positive, negative, or neutral"
    )

    pros: list[str] = Field(
        description="List of positive points mentioned in the review"
    )

    cons: list[str] = Field(
        description="List of negative points mentioned in the review"
    )

    good_things: list[str] = Field(
        description="List of features or aspects that the customer liked"
    )

    rating: int = Field(
        ge=1,
        le=5,
        description="Estimated rating from 1 to 5 based on the review"
    )


model = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash"
)

structured_model = model.with_structured_output(Review)

review = """
I have been using this smartphone for about three weeks, and my overall experience has been quite positive. The camera quality is excellent, especially in daylight, and the pictures come out sharp with accurate colors. The battery life is one of the best features of this phone, as it easily lasts an entire day even with heavy usage. The display is bright, vibrant, and enjoyable for watching videos and playing games.

The performance is smooth for everyday tasks such as browsing, social media, video streaming, and multitasking. Applications open quickly, and I have not experienced any major lag. The design of the phone also feels premium, and it is comfortable to hold.

However, there are a few drawbacks. The device tends to become warm during long gaming sessions and while charging. The charging speed is acceptable but slower than some competing phones in the same price range. I also noticed that the camera performance in low-light conditions could be improved, as images sometimes appear noisy.

Despite these issues, I believe the phone offers great value for money. It provides a good balance of performance, battery life, display quality, and camera features. I would recommend it to students, professionals, and anyone looking for a reliable smartphone without spending too much money.
"""

result = structured_model.invoke(review)

print(result)
print(type(result))
````

````python
from langchain_google_genai import ChatGoogleGenerativeAI

review_schema = {
    "title": "Review",
    "type": "object",
    "properties": {
        "summary": {
            "type": "string",
            "description": "A short summary of the review"
        },
        "sentiment": {
            "type": "string",
            "description": "Overall sentiment: positive, negative, or neutral"
        },
        "pros": {
            "type": "array",
            "items": {"type": "string"},
            "description": "List of positive points mentioned in the review"
        },
        "cons": {
            "type": "array",
            "items": {"type": "string"},
            "description": "List of negative points mentioned in the review"
        },
        "good_things": {
            "type": "array",
            "items": {"type": "string"},
            "description": "List of features or aspects that the customer liked"
        },
        "rating": {
            "type": "integer",
            "description": "Estimated rating from 1 to 5 based on the review"
        }
    },
    "required": [
        "summary",
        "sentiment",
        "pros",
        "cons",
        "good_things",
        "rating"
    ]
}

model = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash"
)

structured_model = model.with_structured_output(review_schema)
review = """
I have been using this smartphone for about three weeks, and my overall experience has been quite positive. The camera quality is excellent, especially in daylight, and the pictures come out sharp with accurate colors. The battery life is one of the best features of this phone, as it easily lasts an entire day even with heavy usage. The display is bright, vibrant, and enjoyable for watching videos and playing games.

The performance is smooth for everyday tasks such as browsing, social media, video streaming, and multitasking. Applications open quickly, and I have not experienced any major lag. The design of the phone also feels premium, and it is comfortable to hold.

However, there are a few drawbacks. The device tends to become warm during long gaming sessions and while charging. The charging speed is acceptable but slower than some competing phones in the same price range. I also noticed that the camera performance in low-light conditions could be improved, as images sometimes appear noisy.

Despite these issues, I believe the phone offers great value for money. It provides a good balance of performance, battery life, display quality, and camera features. I would recommend it to students, professionals, and anyone looking for a reliable smartphone without spending too much money.
"""

result = structured_model.invoke(review)

print(result)  #the output will be dictionary
````
