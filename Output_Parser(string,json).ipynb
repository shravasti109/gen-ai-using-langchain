````python
from dotenv import load_dotenv
from langchain_groq import ChatGroq
from langchain_core.prompts import PromptTemplate

load_dotenv()

llm = ChatGroq(
    model="llama-3.1-8b-instant",
    api_key=""
)

template1 = PromptTemplate(
    template="Write a detailed report on {topic}",
    input_variables=["topic"]
)

template2 = PromptTemplate(
    template="Write a 5 line summary on the following report: {text}",
    input_variables=["text"]
)

prompt1 = template1.invoke({"topic": "planet in solar system"})
result = llm.invoke(prompt1)

prompt2 = template2.invoke({"text": result.content})
result1 = llm.invoke(prompt2)

print(result1.content)
````

````python
from dotenv import load_dotenv
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser

load_dotenv()

from langchain_google_genai import ChatGoogleGenerativeAI

model = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash",
    google_api_key=""
)

template1 = PromptTemplate(
    template="Write a detailed report on {topic}",
    input_variables=["topic"]
)

template2 = PromptTemplate(
    template="Write a 5 line summary on the following report: {text}",
    input_variables=["text"]
)

parser = StrOutputParser()

chain = template1 | model | parser | template2 | model | parser

result = chain.invoke({"topic": "planet in solar system"})

print(result)
````

````python
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import JsonOutputParser

parser = JsonOutputParser()

template = PromptTemplate(
    template="Give me the name, age, and city of a fictional person.\n{format_instruction}",
    input_variables=[],
    partial_variables={
        "format_instruction": parser.get_format_instructions()
    }
)

prompt = template.format()

print(prompt)
#whenever we use parser (json and other two) we add additional variable Partial_variable
#these instructional are told by .get_formal_instruction()
#these is run before run time

#no need to send anything in format because there  is no input yo send

#result = llm.invoke(prompt)


#in output the prompt say Return a JSON object  where did this comes from it is from json output parser
r=model.invoke(prompt)

print(r)
rr=parser.parse(r.content)
print(rr)
````

````python
print(type(rr)) #json is treated as dictionary u cant extracted the information for the dictionar
print("name of the person: ",rr['name'])
print("age of the person:",rr['age'])
print("city where the person live :",rr['city'])
````
