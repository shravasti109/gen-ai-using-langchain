```python
from google_auth_oauthlib.flow import InstalledAppFlow
from google.oauth2.credentials import Credentials
from google.auth.transport.requests import Request
import os

SCOPES = ["https://www.googleapis.com/auth/calendar"]
creds = None

if os.path.exists("token.json"):
    creds = Credentials.from_authorized_user_file("token.json", SCOPES)

if not creds or not creds.valid:
    if creds and creds.expired and creds.refresh_token:
        creds.refresh(Request())
    else:
        flow = InstalledAppFlow.from_client_secrets_file("credentials.json", SCOPES)
        flow.redirect_uri = "urn:ietf:wg:oauth:2.0:oob"
        auth_url, _ = flow.authorization_url(prompt="consent")
        print(f"Go to this URL and authorize:\n{auth_url}\n")
        code = input("Paste the authorization code here: ")
        flow.fetch_token(code=code)
        creds = flow.credentials
    with open("token.json", "w") as token:
        token.write(creds.to_json())
```

```python
from langchain_groq import ChatGroq
from google.colab import userdata  # set via 🔑 icon in Colab sidebar

llm = ChatGroq(
    model="llama-3.3-70b-versatile",
    api_key="........."

)
llm_with_tools = llm.bind_tools(tool)

# ---------------- HELPER: current date/timezone context ----------------
from datetime import datetime
import pytz

def get_system_prompt():
    tz = pytz.timezone("Asia/Kolkata")
    today_str = datetime.now(tz).strftime("%Y-%m-%d")
    year = datetime.now(tz).year
    return f"""You are a scheduling assistant that creates Google Calendar events from natural language.

Today's date is {today_str}. The user's timezone is Asia/Kolkata.

Rules:
- Always resolve relative or partial dates (e.g. "9 march", "tomorrow", "next monday") into a full real date using YYYY-MM-DD format. If no year is mentioned, assume {year}, or {year+1} if that date has already passed this year.
- start_datetime and end_datetime MUST be in the exact format: YYYY-MM-DD HH:MM:SS
- If no time is mentioned (e.g. for a birthday), default to an all-day-style event: 09:00:00 to 10:00:00.
- If no duration is mentioned, default to 1 hour.
- Always set timezone to "Asia/Kolkata" unless told otherwise.
- Write a clear "summary" (title) based on what the user said, e.g. "Shruti's Birthday".
"""

# ---------------- MAIN: type natural language, it books the event ----------------
def book_event(user_text):
    response = llm_with_tools.invoke([
        ("system", get_system_prompt()),
        ("human", user_text)
    ])

    if not response.tool_calls:
        print("Couldn't understand a schedulable event from that. Try rephrasing.")
        return

    for tool_call in response.tool_calls:
        tool_name = tool_call["name"]
        tool_args = tool_call["args"]
        print("Extracted:", tool_args)   # always check before trusting it

        if tool_name in tool_map:
            try:
                result = tool_map[tool_name].invoke(tool_args)
                print(result)
            except Exception as e:
                print(f"Tool call failed: {e}")
        else:
            print(f"Tool '{tool_name}' not found in tool_map")

# ---------------- Try it ----------------
user_input = input("What would you like to schedule? ")
book_event(user_input)

```
